```python
import requests

# define location API
def google_geocode_location(location):
    url = 'http://maps.googleapis.com/maps/api/geocode/json'
    params = {'address' : location}
    response = requests.get(url, params)
    location_data = response.json()
    coords = location_data['results'][0]['geometry']['location']
    return coords

# define weather API
def current_weather_from_darksky(coords):
    url = "https://api.darksky.net/forecast/33ce3d63c8acced80a174cfae548d936/%f,%f" %(coords['lat'], coords['lng'])
    response = requests.get(url)
    weather_data = response.json()
    summary = weather_data['currently']['summary']
    temp = weather_data['currently']['temperature']
    conditions = { 'temp' : temp, 'summary' : summary}
    return conditions
```
