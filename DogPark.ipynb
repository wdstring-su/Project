```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url = ('https://www.topdogparks.com/')
response = requests.get(url)
if response.ok:
    data = response.text
soup = BeautifulSoup(data, 'lxml')
```

```python
name = soup.find_all('h2')
print(name,'\n')
print(name[0].text)

```

```python
features = soup.find_all('div', class_='features')
print(features)
print(features[0].text)
```
